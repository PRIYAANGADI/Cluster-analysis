```python
!pip install scikit-learn
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0) # random state =0 always get the same split every run.
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
iris_data= pd.DataFrame(X_train, columns=iris.feature_names) # label the columns using the strings in iris_dataset.feature_names
```

```python
!pip install mglearn
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import mglearn

flower = scatter_matrix(
    iris_data,
    c=y_train,
    figsize=(10, 10),
    marker='o',
    hist_kwds={'bins': 20},
    s=60,
    alpha=0.8,
    cmap=mglearn.cm3   # now works if mglearn is installed
)

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
```

```python
knn.fit(X_train, y_train) #KNN is like a memory box.remembers all the flowers you showed it.
```

```python
import numpy as np
X_new = np.array([[5, 2.9, 5, 0.2]])
prediction = knn.predict(X_new)


```

```python
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris.target_names[prediction]))


```

```python
y_pred = knn.predict(X_test)
```

```python
print("Test score {:.2f}".format(knn.score(X_test, y_test)))
```
